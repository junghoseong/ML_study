## 3-3 컨볼루셔널 네트워크를 활용한 이미지 분류
### 3-3-1 컨볼루셔널 네트워크를 활용한 이미지 분류 개요
`Convolutional Neural Network`를 사용해 이미지 분류를 수행한다.
### 3-2-2 사진 데이터셋 'CIFAR-10'
10종류의 이미지와 정답 레이블을 모아 둔 데이터셋으로 훈련 데이터 50,000개, 테스트 데이터 10,000개가 포함되어 있다. <br>
이미지는 RGB 3채널 컬러 이미지로 사이즈는 32X32 픽셀이다.
### 3-2-3 CNN이란?
`CNN`은 컨볼루셔널 레이어를 사용해 특징을 추출하는 뉴럴 네트워크로 이미지 인식 분야에서 높은 성능을 발휘한다. <br>
입력 이미지의 특징을 Convolutional layer에서 유지하면서 큰 폭으로 압축하며, pooling layer에서 이미지의 국소적인 왜곡이나 평행 이동의 영향을 쉽게 받지 않도록 robustness를 확보한다. 

- Convolution Layer

입력 데이터의 특징을 표현한 '특징 맵'으로 변환한다. 이미지의 경우에는 사이즈를 줄이면서도 이후 복원할 수 있도록 특징만을 남기는 zip 압축과 유사하다. <br>

> `커널`은 특징을 표시하는 가중치 파라미터의 배열이다. 
> `스트라이드`는 커널을 슬라이드시키는 폭으로, 임의로 조절 가능하다.

컨볼루션을 수행하면 출력되는 특징 맵의 크기가 입력의 크기보다 작아진다. 컨볼루션을 반복 수행할 때에는 패딩(padding)으로 이미지의 주변에 0을 추가하여 원래 이미지의 크기로 복원한다. <br>
<br>
커널 수(필터 수), 커널 사이즈, 스트라이드, 패딩 적용 여부가 Convolutional layer에서의 파라미터이다.

- Pooling Layer

컨볼루셔널 레이어의 출력인 '특징 맵'을 압축하여 데이터 양을 줄이는 레이어다. 부분 구간의 최댓값을 이용하거나 평균값을 사용해 데이터를 압축한다. <br>
<br>
풀링 적용 영역 사이즈, 스트라이드, 패딩 적용 여부는 pooling layer에서의 파라미터이다.

### 3-2-4 패키지 임포트

````python
# 패키지 임포트
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.layers import Activation, Dense, Dropout, Conv2D, Flatten, MaxPool2D
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

### 3-3-5 데이터셋 준비 및 확인
- 데이터셋 준비

`cifar10.load_data()`를 활용해 데이터셋 CIFAR-10을 4종류의 배열에 로드한다.

- 데이터 형태 확인

훈련 데이터와 훈련 라벨은 50,000개, 테스트 이미지와 테스트 라벨은 10,000개이며 RGB 이미지 크기는 32X32X3이다.

- 데이터 셋 이미지 및 라벨 확인

훈련 데이터의 가장 첫 10개의 이미지와 라벨을 확인한다.

````python
# 데이터 셋 준비
(train_images, train_labels), (test_images, test_labels) = cifar10.load_data()

# 데이터 셋 형태 확인
print(train_images.shape)
print(train_labels.shape)
print(test_images.shape)
print(test_labels.shape)

# 데이터 셋 이미지 및 라벨 확인
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(train_images[i])
plt.show()

print(train_labels[0:10])
````

### 3-3-6 데이터셋 전처리 및 확인

- 데이터셋 이미지 전처리

훈련 이미지와 테스트 이미지를 표준화한다. `0~255` 사이인 RGB값을 `0~1`로 변환한다. <br>
<br>
Convolutional layer의 입력이 1차원이던 FC layer와는 다르게 3차원이기 때문에, 이미지 배열의 엘리먼트인 이미지는 1차원이 아니라 3차원을 그대로 사용한다. <br>

- 데이터셋 라벨 전처리

훈련 및 테스트 라벨 배열의 엘리먼트를 one-hot encoding으로 변환한다.

````python
# 데이터셋 이미지 전처리
train_images = train_images.astype('float32') / 255.0
test_images = test_images.astype('float32') / 255.0

# 데이터셋 이미지 전처리 후 형태 확인
print(train_images.shape)
print(test_images.shape)

# 데이터셋 라벨 전처리
train_labels = to_categorical(train_labels, 10)
test_labels = to_categorical(test_labels, 10)

# 데이터셋 라벨 전처리 후 형태 확인
print(train_labels.shape)
print(test_labels.shape)
````

### 3-3-7 모델 생성
컨볼루셔널 뉴럴 네트워크 모델을 생성한다. <br>
`convolutional block` 2개를 겹처서 사용하면서 특징을 추출한다. 이후 `Flatten`을 사용해 3차원 출력을 1차원으로 변환한다. <br>마지막으로 FC layer 2개 겹치고, 이 단계에서 분류를 수행한다.

````python
# 모델 생성
model = Sequential()

#Conv->Conv->Pool->Dropout
model.add(Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=(32, 32, 3)))
model.add(Conv2D(32, (3, 3), activation='relu', padding='same'))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Dropout(0.25))

#Conv->Conv->Pool->Dropout
model.add(Conv2D(64, (3, 3), activation='relu', padding='same', input_shape=(32, 32, 3)))
model.add(Conv2D(64, (3, 3), activation='relu', padding='same'))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Dropout(0.25))

#Flatten -> Dense -> Dropout -> Dense

#Conv->Conv->Pool->Dropout
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(10, activation='softmax'))
````

### 3-3-8 컴파일
손실 함수는 `categorical_crossentropy`, 최적화 함수는 `Adam`, 평가 지표는 `acc`를 활용한다.

### 3-3-9 학습
훈련 이미지와 훈련 라벨의 배열을 모델에 입력해 학습을 시작한다.

````python
# 컴파일
model.compile(loss='categorical_crossentropy', optimizer=Adam(lr=0.001), metrics=['acc'])

# 학습
history = model.fit(train_images, train_labels, batch_size=128, epochs=20, validation_split=0.1)
````

### 3-3-10 모델 저장과 로드
모델을 파일에 저장하려면 model의 `save()` 메서드를 이용한다. 파일에서 모델을 로드할 때는 `load_model()` 메서드를 사용한다.

### 3-3-11 그래프 표시
`fit()`의 반환값인 `history`를 matplotlib을 사용해 그래프에 표시한다.

### 3-3-12 평가
테스트 이미지와 테스트 라벨 배열을 모델에 입력해 평가를 실행해서 정답률을 얻는다.

### 3-3-13 추론
테스트 이미지 중 첫 10개로 추론을 수행해 예측 결과를 얻는다.

````python
# 모델 저장
model.save('convolution.h5')

# 모델 로드
model = load_model('convolution.h5')

# 그래프 표시
plt.plot(history.history['acc'], label='acc')
plt.plot(history.history['val_acc'], label='val_acc')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.legend(loc='best')
plt.show()

# 평가
test_loss, test_acc = model.evaluate(test_images, test_labels)
print('loss: {:.3f}\nacc : {:.3f}'.format(test_loss, test_acc))
````

````python
# 추론할 이미지 표시
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(test_images[i])
plt.show()

# 추론할 라벨 표시
test_predictions = model.predict(test_images[0:10])
test_predictions = np.argmax(test_predictions, axis=1)
labels = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
print([labels[n] for n in test_predictions])
````

### 3-3-14 TPU 사용
Google Colab등 TPU 사용환경에서 참고할 사항을 요약해 본다. TF 1.x 버전 기준이다. <br>
패키지는 keras가 아니라 `tensorflow.keras`를 이용한다. <br>
`compile()`이나 `fit()`은 TPU 모델로 변환한 뒤 실행해야 한다. `keras_to_tpu_model()`을 사용한다. <br>
구체적으로
```python
# TPU 모델로 변환
import tensorflow as tf
import os
tpu_model = tf.contrib.tpu.keras_to_tpu_model(
    model,
    strategy=tf.contrib.tpu.TPUDistributionStrategy(
        tf.contrib.cluster_resolver.TPUClusterResolver(
            tpu='grpc://' + os.environ['COLAB_TPU_ADDR']
        )
    )
)
# 컴파일
tpu_model.compile(loss='categorical_crossentropy', optimizer=Adam(lr=0.001), metrics=['acc'])

# 학습
history = tpu_model.fit(train_images, train_labels, batch_size=128, epochs=20, validation_split=0.1)

# 모델 저장
tpu_model.save('convolution.h5')

# 평가
test_loss, test_acc = tpu_model.evaluate(test_images, test_labels)
print('loss: {:.3f}\nacc : {:.3f}'.format(test_loss, test_acc))

# 추론할 이미지 표시
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(test_images[i])
plt.show()

# 추론할 라벨 표시
test_predictions = tpu_model.predict(test_images[0:10])
test_predictions = np.argmax(test_predictions, axis=1)
labels = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
print([labels[n] for n in test_predcitions])
```



TPU를 사용하는 학습, 평가, 추론에서 지정하는 학습 데이터 수는 TPU 코어 수로 나누어떨어지도록 설정해야한다.<br> 그렇지 않으면 학습 데이터 수를 늘려서 추론한 뒤, 앞쪽 몇개만 이용한다.
