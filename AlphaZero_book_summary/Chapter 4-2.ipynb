### 4-2-1 정책 경사법을 활용한 미로 게임 풀기
정책에 따라 행동하면서 성공 시의 행동은 중요한 것으로 판단하고, 그 행동을 많이 할 수 있도록 정책을 갱신하는 방법을 `정책 경사법`이라고 부른다. <br>
에이전트가 상하좌우로 이동해 우측 하단의 골인 지점까지 도착하는 것을 목적으로 정책 경사법을 활용해 미로 게임을 공략한다.

### 4-2-2 정책 경사법 학습 순서
1. 파라미터 $\theta$ 준비
2. 파라미터 $\theta$를 정책으로 변환
3. 정책에 따라 골인 지점에 이를 때까지 행동 반복
4. 성공한 행동을 많이 받아들이도록 파라미터 $\theta$를 변경
5. 정책 변화량이 임곗값 이하가 될 때까지 2~4 단계를 반복한다.

- 정책 

정책은 '특정한 상태에서 특정한 행동을 할 확률'을 의미한다. 주로 함수나 뉴럴 네트워크로 표시하나 이 task에서는 '테이블 형식'으로도 충분하다. <br>
상태별로 행동을 선택할 확률을 '상태 수 X 행동 수' 배열로 정책을 저장한다.

- parameter $\theta$

파라미터 $\theta$는 정책으로 변환되는 값으로 딥 러닝에서 가중치 파라미터에 해당한다.

### 4-2-5 파라미터 $\theta$의 초기값 준비
파라미터 $\theta$의 초기값을 준비한다. 학습 전에는 올바른 파라미터를 알지 못하므로, 이동할 수 있는 방향은 1, 이동할 수 없는 방향은 `np.nan`(Numpy에서의 결손값)으로 초기화한다.

```python
theta_0 = np.array([
    [np.nan, 1, 1, np.nan], # 0 상, 우, 하, 좌
    [np.nan, 1, 1, 1], # 1
    [np.nan, np.nan, np.nan, 1], # 2
    [1, np.nan, 1, np.nan], # 3
    [1, 1, np.nan, np.nan], # 4
    [np.nan, np.nan, 1, 1], # 5
    [1, 1, np.nan, np.nan], # 6
    [np.nan, np.nan, np.nan, 1], # 7
])
```

### 4-2-6 파라미터 $\theta$의 정책으로 변화 & 4-2-7 $\theta$의 초기값을 정책으로 변환
정책 경사법에서는 'softmax function'을 이용하여 파라미터 $\theta$를 정책으로 변환한다. Softmax 함수는 여러 값의 합계가 1이 되도록 해당 값들을 0~1 사이의 실수값으로 변경한다.
$$ [np.nan, 1, 1, np.nan] -> [0, 0.5, 0.5, 0]$$
이 함수는 딥 러닝의 분류 출력 레이어에서도 활용되며, 파라미터 $\theta$가 음수라도 계산 가능한 이점이 있다. 수식은 다음과 같다.
$$f(x_i) = \frac{e^{\theta_i}}{\displaystyle\sum_{j=1}^{n}e^{\theta_j}}$$

```python
# 파라미터 theta를 정책으로 변환
def get_pi(theta):
    # 소프트맥스 함수로 변환
    [m, n] = theta.shape
    pi = np.zeros((m, n))
    exp_theta = np.exp(theta)
    
    for i in range(0, m):
        pi[i, :] = exp_theta[i, :] / np.nansum(exp_theta[i, :])
    
    pi = np.nan_to_num(pi)
    return pi

# 파라미터 theta의 초기값을 정책으로 변환
pi_0 = get_pi(theta_0)
print(pi_0)
```

### 4-2-8 정책에 따라 행동 얻기 & 4-2-9 행동에 따라 다음 상태 얻기
정책에 따라 행동을 얻는 함수를 생성한다. `np.random.choice()`는 인수 `p`의 확률 분포에 따라 배열의 엘리먼트를 무작위로 반환하는 함수다. <br>
p에 임의의 상태의 정책, 즉 임의의 확률 분포를 지정한다. <br>
<br>
행동에 따라 다음 상태를 얻는 함수를 생성한다. 좌우 이동은 1, 상하 이동은 3만큼 이동한다.

```python
# 정책에 따라 행동 얻기
def get_a(pi, s):
    # 정책의 확률에 따라 행동 반환
    return np.random.choice([0, 1, 2, 3], p=pi[s])

# 행동에 따라 다음 상태 얻기
def get_s_next(s, a):
    if a == 0: # 상
        return s - 3
    elif a == 1: # 우
        return s + 1
    elif a == 2: # 하
        return s + 3
    elif a == 3: # 좌
        return s - 1
```

### 4-2-10 에피소드 실행, 4-2-11 이력 확인, 4-2-12 파라미터 $\theta$ 갱신
1 에피소드를 실행하고 [상태, 행동]의 리스트의 이력을 얻는다. 실행 결과를 확인하면, 골인까지 어떤 경로로 얼마의 스텝이 걸렸는지 알 수 있다.<br>
<br>
정책 경사법에서 파라미터 $\theta$의 변경식은 다음과 같다. 파라미터 $\theta$에 학습률과 파라미터 $\theta$의 변화량을 곱한 뒤 더한다.
$$ \theta(s, a) \leftarrow \theta(s, a) + \eta \times \frac{N(s, a) - P(s, a) \times N(s)}{T}$$
T : 골인 지점까지 걸린 총 횟수 <br>
N(s, a) : 상태 s에서 행동 a를 선택한 횟수 <br>
N(s) : 상태 s에서 무엇인가 행동을 선택한 횟수 <br>
P(s, a) : 상태 s에서 행동 a를 선택하는 정책 

```python
# 1 에피소드 실행 후 이력 취득
def play(pi):
    s = 0 # 상태
    s_a_history = [[0, np.nan]] # 상태와 행동 이력
    
    # 에피소드 종료 시까지 반복
    while True:
        # 정책에 따라 행동 얻기
        a = get_a(pi, s)
        
        # 행동에 따라 다음 상태 얻기
        s_next = get_s_next(s, a)
        
        # 이력 갱신
        s_a_history[-1][1] = a
        s_a_history.append([s_next, np.nan])
        
        # 종료 판정
        if s_next == 8:
            break
        else:
            s = s_next
            
    return s_a_history

# 1 에피소드 실행 및 이력 확인
s_a_history = play(pi_0)
print(s_a_history)
print('1 에피소드의 스텝 수: {}'.format(len(s_a_history)+1))
```

```python
# 파라미터 theta 갱신
def update_theta(theta, pi, s_a_history):
    eta = 0.1 # 학습 계수
    total = len(s_a_history) - 1 # 골인 지점까지 걸린 총 스텝 수
    [s_count, a_count] = theta.shape # 상태 수, 행동
    
    # 파라미터 theta의 변화량 계산
    delta_theta = theta.copy()
    for i in range(0, s_count):
        for j in range(0, a_count):
            if not(np.isnan(theta[i, j])):
                # 상태 s에서 행동 a를 선택한 횟수
                sa_ij = [sa for sa in s_a_history if sa == [i, j]]
                n_ij = len(sa_ij)
                
                # 상태 s에서 무엇인가 행동을 선택한 횟수
                sa_i = [sa for sa in s_a_history if sa[0] == i]
                n_i = len(sa_i)
                
                # 파라미터 theta의 변화량
                delta_theta[i, j] = (n_ij - pi[i, j] * n_i) / total
                
    # 파라미터 theta 갱신
    return theta + eta * delta_theta
```

### 4-2-13 에피소드를 반복해서 실행하며 학습하기 & 4-2-14 학습 실행 결과 확인

```python
stop_epsilon = 10 ** -4 # 역치
theta = theta_0 # 파라미터 theta
pi = pi_0 # 정책

# 에피소드를 반복해서 실행하며 학습
for episode in range(10000):
    # 1 에피소드 실행 후 이력 얻기
    s_a_history = play(pi)
    
    # 파라미터 theta 갱신
    theta = update_theta(theta, pi, s_a_history)
    
    # 정책 갱신
    pi_new = get_pi(theta)
    
    # 정책 변화량
    pi_delta = np.sum(np.abs(pi_new - pi))
    pi = pi_new
    
    # 출력
    print("에피소드: {}, 스텝: {}, 정책 변화량: {:.4f}".format(episode, len(s_a_history) - 1, pi_delta))
          
    # 완료 판정
    if pi_delta < stop_epsilon: # 정책 변화량이 임계값 이하
        break
```

### 4-2-15 애니메이션 표시
-`animate()`

`animate()`는 애니메이션 프레임별 처리를 수행하는 함수로 인수 `i`는 몇 번째 프레임인지를 의미한다. `Line2D`에서 제공하는 `circle`의 `set_data()`로 xy좌표를 갱신한다.

- `animate.FuncAnimation`

`animate.FuncAnimation`은 애니메이션을 생성하는 클래스다. 인수로는 `Figure`, 애니메이션의 정기적 처리를 수행하는 함수, 최대 프레임 수, 인터벌, 반복여부를 지정한다. <br>
`to_jshtml()`를 이용하여 HTML로 변환한다.

```python
# 애니메이션 정기 처리를 수행하는 함수
def animate(i):
    state = s_a_history[i][0]
    circle.set_data((state % 3) + 0.5, 2.5 - int(state / 3))
    return circle

# 애니메이션 표시
anim = animation.FuncAnimation(fig, animate,
                              frames = len(s_a_history), interval=200, repeat=False)
HTML(anim.to_jshtml())
```
